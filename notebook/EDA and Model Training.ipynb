# Data Standardization and Label Encoding

```python
numerical_categories = X.select_dtypes(exclude='object').columns
cat_categories = X.select_dtypes(include='object').columns

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

numerical_transformer = StandardScaler()
categorial_transformer = OneHotEncoder(handle_unknown='infrequent_if_exist',
min_frequency=5,sparse_output=False)

preprocessor = ColumnTransformer([
    ("Numerical transformer",numerical_transformer,numerical_categories),
    ('categorial transformer',categorial_transformer,cat_categories)
],remainder='drop'
)
```

# PipeLine creation


```python
from sklearn.pipeline import Pipeline
from lightgbm import LGBMRegressor

model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ('Random Forest Regressor', LGBMRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=8,
    random_state=42,
    verbosity=-1)),
])
```

# Model Training

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred[0]
```

```python
from sklearn.metrics import r2_score, mean_squared_error, root_mean_squared_error

print("R2_score: ", r2_score(y_test, y_pred))
print("mean_squared_error: ", mean_squared_error(y_test, y_pred))
print("root_mean_squared_error: ", root_mean_squared_error(y_test, y_pred))
```

# Testing on test.csv

```python
import pandas as pd
import numpy as np

test_df = pd.read_csv("data/test.csv")


X_test_kaggle = test_df.drop(columns=['Id'], errors='ignore')

y_pred_raw = model.predict(X_test_kaggle)

y_pred = np.expm1(y_pred_raw) if np.any(y_pred_raw < 0) or y_pred_raw.mean() < 10 else y_pred_raw

y_pred = np.clip(y_pred, 0, None)

submission = pd.DataFrame({
    'Id': test_df['Id'],
    'SalePrice': y_pred
})

submission.to_csv("submission.csv", index=False)

print("Submission file created: submission.csv")
print(submission.head())
```
